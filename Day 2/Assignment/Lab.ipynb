## 1. Students Dictionary

```python
students = {
    "name":["Omar","yousef"],
    "salary":[7500,5000]
}
```

### Add new students

```python
students["name"].append("ahmed")
students["salary"].append(8000)
students["name"].append("ali")
students["salary"].append(10000)
```

### Print all students using f-string

```python
for i in range(len(students["name"])):
    print(f"Name: {students['name'][i]}, Salary: {students['salary'][i]})
```

### Print using format

```python
for i in range(4):
    print("Name:{},Salary:{}".format(students["name"][i],students["salary"][i]))
```

## 4. Class Emp

```python
class Emp:
    data = {}

    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary

        Emp.data[name] = {
            "age": age,
            "salary": salary
        }

    def add(self, name, age, salary):
        Emp.data[name] = {
            "age": age,
            "salary": salary
        }

    def display():
        print(Emp.data)

e1 = Emp("Omar", 25, 15000)
e1.add("Ali", 30, 20000)

print(Emp.data)
```
