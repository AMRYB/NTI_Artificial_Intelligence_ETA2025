```
import pandas as pd

df = pd.read_csv("salaries.csv")
df.head()
```

```

df.isnull().sum()
cat_cols = [
    "experience_level", "employment_type", "job_title",
    "salary_currency", "employee_residence",
    "company_location", "company_size"
]

for col in cat_cols:
    df[col] = df[col].astype("category")

df.info()
```

```
Q1 = df['salary_in_usd'].quantile(0.25)
Q3 = df['salary_in_usd'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['salary_in_usd'] < Q1 - 1.5 * IQR) |
              (df['salary_in_usd'] > Q3 + 1.5 * IQR)]
outliers.head()
df['remote_ratio'].unique()

```

```
df = df[df['remote_ratio'].isin([0, 50, 100])]
df.info()
df = df[(df['work_year'] >= 2019) & (df['work_year'] <= 2025)]
df.to_csv("cleaned_file.csv", index=False)
```

```
import pandas as pd
import os

df = pd.read_csv("/content/salaries.csv")
print("أول 5 صفوف من الملف الأصلي:") # First 5 rows of the original file
df.head()
```

```
print("التحقق من القيم المفقودة:") # Checking for missing values
df.isnull().sum()

cat_cols = [
    "experience_level", "employment_type", "job_title",
    "salary_currency", "employee_residence",
    "company_location", "company_size"
]

for col in cat_cols:
    df[col] = df[col].astype("category")

print("معلومات DataFrame بعد تحويل الأعمدة الفئوية:") # DataFrame info after converting categorical columns
df.info()
```

```
Q1 = df['salary_in_usd'].quantile(0.25)
Q3 = df['salary_in_usd'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['salary_in_usd'] < Q1 - 1.5 * IQR) |
              (df['salary_in_usd'] > Q3 + 1.5 * IQR)]
print("القيم المتطرفة في عمود salary_in_usd:") # Outliers in the 'salary_in_usd' column
outliers.head()

print("القيم الفريدة في عمود remote_ratio:") # Unique values in the 'remote_ratio' column
df['remote_ratio'].unique()
```

```
df = df[df['remote_ratio'].isin([0, 50, 100])]
print("معلومات DataFrame بعد تصفية remote_ratio:") # DataFrame info after filtering 'remote_ratio'
df.info()
df = df[(df['work_year'] >= 2019) & (df['work_year'] <= 2025)]
df.to_csv("cleaned_file.csv", index=False)
print("تم حفظ الملف النظيف في cleaned_file.csv") # Cleaned file saved to 'cleaned_file.csv'
```

```
file_path = '/content/salaries.csv'
file_size_bytes = os.path.getsize(file_path)
file_size_kb = file_size_bytes / 1024
file_size_mb = file_size_kb / 1024

print(f"حجم الملف الأصلي ({file_path}):") # Size of the original file
print(f"{file_size_bytes:.2f} بايت") # Bytes
print(f"{file_size_kb:.2f} كيلوبايت") # Kilobytes
print(f"{file_size_mb:.2f} ميغابايت") # Megabytes
```
